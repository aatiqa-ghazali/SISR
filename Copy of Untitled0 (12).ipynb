```python
!pip install scikit-image
from skimage.metrics import structural_similarity
import cv2
import numpy as np
!pip install scikit-image
from skimage.metrics import structural_similarity
import cv2
imageA='/content/drive/MyDrive/SISR/Pyramids_deraining/pyramids-derain/code/test_img/rain/Manga109/HR/x2/AkkeraKanjinchou_HR_x2.png'
imageB= '/content/drive/MyDrive/SISR/Pyramids_deraining/pyramids-derain/code/test_img/results/manga109/rain/AkkeraKanjinchou_LRBI_x2.png'
im = cv2.imread(imageA)
(H, W,T) = im.shape
imageB=cv2.imread(imageB)
imageB=cv2.resize(np.float32(imageB), (W, H))

from skimage.metrics import mean_squared_error
from skimage.metrics import peak_signal_noise_ratio
img1 = im
img2 = imageB

MSE = mean_squared_error(img1, img2)
PSNR = peak_signal_noise_ratio(img1, img2)
print('MSE: ', MSE)
print('PSNR: ', PSNR)

SSIM = structural_similarity(img1, img2, multichannel=True)
print('SSIM: ', SSIM)
```
