```python
#!/usr/bin/python
from PIL import Image
import os, sys

path = "/content/drive/MyDrive/SISR/Pyramids_deraining/pyramids-derain/code/TrainData/data2017/DIV2K_train_LR_bicubic/X4/"
dirs = os.listdir( path )

def resize():
    for item in dirs:
        if os.path.isfile(path+item):
            im = Image.open(path+item)
            f, e = os.path.splitext(path+item)
            imResize = im.resize((512,512), Image.ANTIALIAS)
            imResize.save(f + ' resized.png', 'png', quality=90)

resize()
```

```python
import cv2
import numpy as np

imageA='/content/drive/MyDrive/SISR/Pyramids_deraining/pyramids-derain/code/test_img/rain/Manga109/HR/x4/ARMS_HR_x4.png'
imageB='/content/drive/MyDrive/SISR/Pyramids_deraining/pyramids-derain/code/test_img/results/manga109/1/ARMS_LRBI_x4.png'
im = cv2.imread(imageA)
(H, W,T) = im.shape
imageB=cv2.imread(imageB)
imageB=cv2.resize(np.float32(imageB), (W, H))

from skimage.metrics import mean_squared_error
from skimage.metrics import peak_signal_noise_ratio
img1 = cv2.imread('/content/drive/MyDrive/SISR/Pyramids_deraining/pyramids-derain/code/test_img/rain/Manga109/HR/x4/ARMS_HR_x4.png')
img2 = imageB

MSE = mean_squared_error(img1, img2)
PSNR = peak_signal_noise_ratio(img1, img2)
print('MSE: ', MSE)
print('PSNR: ', PSNR)
```

```python
import cv2

img1 = cv2.imread(imageA)
img2 = imageB
SSIM = structural_similarity(img1, img2, multichannel=True)
print('SSIM: ', SSIM)
```
